```python
import sys

print('PADHARO CHATOKDA')


print(' MENU CARD ') 
lst_choice=[ ]
lst_qty=[ ]

dict_menu={1:'Italian Pasta', 2:'Mommos', 3:'Thai Noodles',4:'Brownie'}
dict_price={1:200 , 2:150 ,3:250 , 4:250}  

while(1): 
    print('1.Italian Pasta: 200 per plate\n2.Mommos: 150 per plate\n3.Thai Noodles: 250 per plate\n4.Brownie: 250 per unit') 
    try: 
        a=input('Do you want to palce order(y/n)?')
    except: 
        print('Oops!',sys.exc_info()[0],'occured')
        continue 
        
    if a=='y' or a=='Y': 
        try: 
            choice=int(input('Enter Your Choice.'))
        except: 
            print('Oops!',sys.exc_info()[0],'occured')
            continue   
        
        if choice>=1 and choice<=4: 
            try: 
                qty=int(input('How many do you need?'))
            except: 
                print('Oops!',sys.exc_info()[0],'occured')
                continue   
        
            if qty>=1: 
                lst_choice.append(choice)
                lst_qty.append(qty)
            else: 
                print('Wrong Input\n')
                continue
        else:  
            print('Wrong Input\n')
            continue
    elif a=='n' or a=='N': 
        print('\nBill:')
        total=0 
        length=len(lst_choice)
        for i in range(0, length): 
            name=lst_choice[i]
            print('item=',dict_menu[name])
            print('Qty: ',lst_qty[i])
            print('Rs: ',(dict_price[name]),'per unit')
            print('\n')               
            total = total + (dict_price[name] * lst_qty[i])
        print('Total: ',total)
        print('Thank You! Visit Again.') 
        break
    
    else: 
        print('Wrong Input\n')
        continue
        
        
```
